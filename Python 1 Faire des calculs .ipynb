<b>Un exemple en chimie : Calcul du réactif limitant et bilan de matière.</b>

```python
## Bilan de matière
print("Équation d'une réaction considérée comme totale : aA + bB -> cC + dD")
print ("Entrer les valeurs des nombres stoechiometriques a, b, c et d")
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))
d = float(input("d = "))
print("Entrer la quantité initiale de  réactifs  A et B en mol")
nA = float(input("n0(A) = "))
nB = float(input("n0(B) = "))
xmaxA = nA / a
xmaxB = nB / b         # On calcule les avancements maximum possibles
if xmaxA == xmaxB:     # On compare les 2 possibilités d'avancement maximum obtenues
    print("Le mélange est stoechiométrique")
    xmax = xmaxA
elif xmaxA < xmaxB:
    print("A est le réactif limitant")
    xmax = xmaxA
else:
    print("B est le réactif limitant")
    xmax = xmaxB
print("xmax =", xmax, "mol")
nfA = nA - a * xmax    # On calcule la quantité de matière finale pour le réactif A
nfB = nB - b * xmax    # On calcule la quantité de matière finale pour le réactif B
nfC = c * xmax         # On calcule la quantité de matière finale pour le produit C
nfD = d * xmax         # On calcule la quantité de matière finale pour le produit D
print("Bilan de matière à l'état final :")
print("nf(A) = {} mol\nnf(B) = {} mol\n nf(C) = {} mol\nnf(D) = {} mol\n".format(nfA, nfB, nfC, nfD))
```

# Boucle for ... Répéter une instruction un nombre de fois déterminé. 

L’ordinateur est très doué pour effectuer des tâches répétitives. Les boucles permettent d’exécuter plusieurs 
fois le même bloc d’instructions. Cela sera particulièrement intéressant pour les activités où il faut parcourir une 
liste et faire le même calcul pour chaque indice.

<div class="alert alert-warning">
<b>Travail à faire : </b>
    Exécuter les codes suivants.

```python
print("début")
for i in range(10):   # i vaudra successivement 0,1,2,3,4,5,6,7,8 et 9 ! Mais pas 10 !
    print(i)
print("fin")          # Cette instruction n'est pas indentée, elle s'exécute donc une seule fois lorsque la boucle est terminée.
```

<div class="alert alert-danger">
<b> Vous remarquerez que l'indice donné (ici 10 ) correspond à la première valeur entière non atteinte. <b><br> Le premier indice est toujours zéro si on ne le précise pas.<br> Ici encore, l'indentation permet de délimiter les instructions qui doivent être répétées.

```python
print("début")
for i in range(3,10):   # On peut aussi préciser le premier indice utilisé et le premier non utilisé.
    print(i)
print("fin")
```

```python
print("début")
for i in range(10,17,2):   # On peut aussi imposer un pas pour les valeurs successives de l'indice.
    print(i)
print("fin")
```
